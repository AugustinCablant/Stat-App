```python
import os
#os.chdir('/work/Stat-App/Notebooks')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA : Temporal analysis

```python
#os.chdir('/Stat-App/DATA')
df_90_95 = pd.read_csv('https://raw.githubusercontent.com/AugustinCablant/Stat-App/enzo/DATA/from_countries_to_region/regions_90_95')
df_95_00 = pd.read_csv('https://raw.githubusercontent.com/AugustinCablant/Stat-App/enzo/DATA/from_countries_to_region/regions_95_00')
df_00_05 = pd.read_csv('https://raw.githubusercontent.com/AugustinCablant/Stat-App/enzo/DATA/from_countries_to_region/regions_00_05')
df_05_10 = pd.read_csv('https://raw.githubusercontent.com/AugustinCablant/Stat-App/enzo/DATA/from_countries_to_region/regions_05_10')
columns = ['Central America ', 'North America', 'South America', 'North Africa', 'Sub-Saharan Africa', 'Northern Europe', 'Western Europe', 
               'Southern Europe', 'Eastern Europe', 'Western Asia', 'South Asia', 'South-East Asia', 'Oceania', 'Central Asia ']
df_90_95.index = columns
df_90_95 = df_90_95[columns] 
df_95_00.index = columns
df_95_00 = df_95_00[columns] 
df_00_05.index = columns
df_00_05 = df_00_05[columns] 
df_05_10.index = columns
df_05_10 = df_05_10[columns] 
df=df_05_10.compare(df_90_95)
df.index=columns
df = df[columns] 
df.head()
```

```python
sns.heatmap(df, cmap="viridis", robust = True)
plt.show()
```

```python
log_data = np.log10(df)
sns.heatmap(log_data, cmap="viridis")
plt.show()
```

```python
for region in columns:
    df1=pd.DataFrame({region:[df_90_95[region].sum(), df_95_00[region].sum(), df_00_05[region].sum(), df_05_10[region].sum()]})
    df1.index= ['1990', '1995', '2000', '2005']
    plt.plot(df1, label=region)
    plt.legend(bbox_to_anchor=(1.0, 1.0, 0.1, 0), loc="upper left")
plt.title('Destinations of migrants', fontsize=20)
```

```python
for region in columns:
    df2=pd.DataFrame({region:[df_90_95.loc[region].sum(axis=0), df_95_00.loc[region].sum(axis=0), df_00_05.loc[region].sum(axis=0), df_05_10.loc[region].sum(axis=0)]})
    df2.index= ['1990', '1995', '2000', '2005']
    plt.plot(df2, label=region)
    plt.legend(bbox_to_anchor=(1.0, 1.0, 0.1, 0), loc="upper left")
plt.title('Origins of migrants', fontsize=20)
```
