```python
def overdetermined(df):
    n=len(df4.values)
    T=np.copy(df.values[:(n-1),:(n-1)])
    n=len(T)
    m = 0
    for i in range(n):
        for j in range(n):
            if T[i][j]!=0:
                m+=1
    return m + 2*n - n**2 - 2 < 0, m + 2*n - n**2 - 2
```
