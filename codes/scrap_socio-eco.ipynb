```python
import os
import bs4
import urllib
from urllib import request
import pandas as pd
```

```python
#Scrapping Pib/hab

url = "https://fr.wikipedia.org/wiki/Liste_des_pays_par_PIB_(PPA)_par_habitant"
    
request_text = request.urlopen(url).read()
page = bs4.BeautifulSoup(request_text, "html.parser")

# Get the sortable table 
table_new = page.find("table", {'class' : 'wikitable alternance'})
#print(table_new)

rows = table_new.find_all('tr')
d = dict()
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        d[cols[0]] = cols[1:]


df = pd.DataFrame.from_dict(d,orient='index')

for row in rows:
    cols = row.find_all('th')
    #print(cols)
    if len(cols) > 0 : 
        cols = [ele.get_text(separator=' ').strip().title() for ele in cols]
        columns_new= cols
columns_new
df.columns = columns_new[1:]

# We keep only the interesting columns
df=df.rename({'Pib Par Habitant [ A ]':'Pib Par Habitant'}, axis=1)

df.head()
```

```python
#Scrapping IDH

url2 = "https://fr.wikipedia.org/wiki/Variations_de_l%27IDH_depuis_1990"
    
request_text2 = request.urlopen(url2).read()
page2 = bs4.BeautifulSoup(request_text2, "html.parser")

# Get the sortable table 
table_new2 = page2.find("table", {"class": "wikitable", "class": "sortable"})
#print(table_new2)

rows2 = table_new2.find_all('tr')
pays=[]
for row in rows2:
    cols2 = row.find_all('span' , {"class" : "datasortkey"})
    pays.append([span.get_text(separator=' ').strip().title() for span in cols2])
pays.pop(0)
pays = pd.DataFrame.from_dict(pays)

d2 = dict()
for row in rows2:
    cols2 = row.find_all('td')
    cols2 = [ele.text.strip() for ele in cols2]
    if len(cols2) > 0 : 
        d2[cols2[0]] = cols2[1:]
        
df2 = pd.DataFrame.from_dict(d2,orient='index')

df2.reset_index(inplace=True)
pays.columns=['Pays']
df2.columns=['1990', '1995', '2000', '2005', '2010', '2015', '2020', '2021']
df2=pays.join(df2)
df2.head()
```

```python
#Scrapping number of inhabitants

url3 = "https://fr.wikipedia.org/wiki/Liste_des_pays_par_population"
    
request_text3 = request.urlopen(url3).read()
page3 = bs4.BeautifulSoup(request_text3, "html.parser")

# Get the sortable table 
table_new3 = page3.find("table", {"class": "wikitable", "class": "sortable", "class": "alternance"})

rows3 = table_new3.find_all('tr')

d3 = dict()
for row in rows3:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        d3[cols[0]] = cols[1:]

df3 = pd.DataFrame.from_dict(d3,orient='index')
df3.drop(labels='-', inplace=True)

df3.head()
```
